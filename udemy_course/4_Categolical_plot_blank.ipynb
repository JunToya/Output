```python
df_adult['age'].mean()
```

'age'と'pclass'の関係

```python
df.groupby('pclass')['fare'].describe()
```

```python
plt.figure(figsize=(12, 5))
sns.histplot(data=df_adult, x='age', bins=13, hue='pclass', palette='Set1', 
             multiple='dodge', shrink=0.8)
plt.title("adultの年齢分布")
plt.show()
```

```python
# figsize : インチ単位
# dpi : ドットパーインチ
# 例） figsize=(10, 5), dpi=100の場合
# 10×100=1000 5×100=500
# 1000×500=50万ピクセル（50万画素）
```

```python
fig, axes = plt.subplots(2, 1, figsize=(15, 10), tight_layout=True, sharex=True)
sns.histplot(data=df_adult, x='age', bins=13, hue='pclass', multiple='dodge',
            palette='Set1', shrink=0.9, ax=axes[0])
sns.histplot(data=df_adult, x='age', bins=13, hue='pclass', multiple='fill',
            palette='Set2', shrink=0.9, ax=axes[1])
axes[1].set_ylabel("Ratio")
axes[0]

fig.suptitle("adultsの年齢分布")

plt.show()
```

'age'と'sex'の関係

```python
plt.figure(figsize=(12, 5))
sns.histplot(data=df_adult, x='age', bins=13, hue='sex', 
             multiple='dodge', shrink=0.8)
plt.title("adultの年齢分布")
plt.show()
```

相関関係

```python
df_adult.corr(numeric_only=True)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df_adult.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('df_adult　相関関係')
plt.show()
```

### キャットプロットとストリッププロット

```python
df.head()
```

```python
df.groupby('embarked')['age'].describe()
```

```python
sns.catplot(data=df, x='embarked', y='age', height=6, aspect=2)
plt.show()
```

```python
sns.catplot(data=df, x='embarked', y='age', kind='box', height=6, aspect=2)
plt.show()
```

```python
sns.catplot(data=df, x='embarked', y='age', kind='violin', height=6, aspect=2)
plt.show()
```

```python
sns.catplot(data=df, x='embarked', y='age', hue='sex', height=6, aspect=2)
plt.show()
```

```python
sns.catplot(data=df, x='embarked', y='age', hue='sex', dodge=True, height=6, aspect=2)
plt.show()
```

```python
sns.catplot(data=df, x='embarked', y='age', height=4, col='pclass', 
            palette='Greens')
plt.show()
```

```python
sns.catplot(data=df, x='embarked', y='age', height=4, col='pclass', 
            row='survived', palette='Reds')
plt.show()
```

```python
sns.catplot(data=df, x='embarked', y='age', height=4, col='pclass', 
            row='sex', hue='survived', palette='Set2')
plt.show()
```

```python
plt.figure(figsize=(6, 3))
sns.stripplot(data=df, x='embarked', y='age', hue='survived', palette='Set2')
plt.show()
```

カウントプロット

```python
sns.countplot(data=df, x='survived', hue='sex', palette='gnuplot')
plt.show()
```

```python
sns.countplot(data=df, x='survived', hue='embarked', palette='bone')
plt.xticks([0, 1], ['No', 'Yes']) # xticksで要素名を変更（0,1 → No,Yes）
plt.show()
```

### バープロット

```python
sns.barplot(data=df, x='embarked', y='age', palette='Set2')
plt.xticks([0, 1, 2], ['Southampton', 'Cherbourg', 'Queenstown'], 
           rotation=20) # rotationで文字列の角度を変更できる
plt.show()
```

barplotはデフォルトでy軸'age'の平均値を表示させる

```python
df.groupby('embarked')['age'].mean()
```

### ボックスプロット

```python
sns.boxplot(data=df, x='pclass', y='fare', hue='who', palette='Set3')
plt.show()
```

```python
sns.boxplot(data=df, x='pclass', y='fare', hue='who', palette='Set3', 
            sym='') #sym=''で外れ値を表示させない
plt.legend(loc='upper center', title='Who')
plt.show()
```

```python
df.groupby(['pclass', 'who'])['fare'].describe()
```

### バイオリンプロット

```python
plt.figure(figsize=(8,6))
sns.violinplot(data=df, x='sex', y='age', hue='survived', palette='spring')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.violinplot(data=df, x='sex', y='age', hue='survived', palette='spring',
              split=True) #hueがニ値の場合、split=Trueで比較しやすくなる
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.violinplot(data=df, x='sex', y='age', hue='survived', palette='spring',
              split=True, inner='quartile') # inner='quartile'で四分位範囲がわかりやすくなる
plt.show()
```

### スウォームプロット

```python
plt.figure(figsize=(10, 6))
sns.swarmplot(data=df, x='sex', y='age', hue='survived', palette='prism')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.swarmplot(data=df, x='sex', y='age', hue='survived', palette='prism',
             dodge=True, size=4)
plt.legend(loc='upper center', title='Survived')
plt.show()
```
