```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold, cross_val_score
from sklearn.metrics import accuracy_score

import lightgbm as lgb
```

```python
df_train = pd.read_csv('../train.csv')
df_test = pd.read_csv('../test.csv')
df_sub = pd.read_csv('../sample_submission.csv')
```

```python
X_train = df_train.drop(['id', 'Target'], axis = 1)
y = df_train['Target']

label_mapping = {'Dropout': 0, 'Enrolled': 1, 'Graduate': 2}
encoded_y = y.map(label_mapping)
```

```python
lgb = lgb.LGBMClassifier(random_state=42)
lgb.fit(X_train, encoded_y)

cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(lgb, X_train, encoded_y, cv=cv_strategy, scoring='accuracy')

print(f'Cross-validation scores: {cv_scores}')
print(f'Average cross-validation score: {cv_scores.mean()}')
print(f'Standard Error: {cv_scores.std()}')
```
